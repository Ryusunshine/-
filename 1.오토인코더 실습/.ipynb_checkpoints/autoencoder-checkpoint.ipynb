### 왜 오토인코더를 사용할까?
이 시점에서, 입력을 왜 예측할 필요가 있는지, 그리고 오토인코더를 어떻게 응용할 수 있는지 의문을 가질 수 있다.

오토인코더의 주요primary 응용 분야는 이상 감지anomaly detection 또는 이미지 노이즈 제거image denoising이다. 오토인코더의 작업이 매니폴드 즉 주어진 데이터 매니폴드에 있는 데이터를 재건하는 것임을 알고 있고, 오토인코더가 해당 매니폴드 안에 존재하는 입력만 재건할 수 있기를 원한다. 따라서 우리는 모델이 훈련 중에 관찰한 것들만을 재건할 수 있도록 제한하고, 따라서 새로운 입력에 존재하는 어떠한 변화variation도 제거되는데, 왜냐하면 이 모델은 이러한 미세한 변화perturbations에 영향을 받지 않을insensitive 것이기 때문이다.

오토인코더의 또 다른 응용은 이미지 압축기compressor이다. 만일 입력 차원 nn 보다 낮은 중간 차원 dd 를 갖고 있다면, 인코더는 압축기로 사용될 수 있고, 은닉 표현hidden representation (코딩 된 표현)은 특정 입력의 모든 (또는 대부분) 정보를 전달하지만 적은 공간을 차지한다.

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
from utils import load_mnist
from trainer import Trainer
```

```python
from argparse import Namespace

config = {
    'train_ratio': .8,
    'batch_size': 256,
    'n_epochs': 50,
    'verbose': 1,
    'btl_size': 2
}

config = Namespace(**config)

print(config)
```

```python
def show_image(x): #자세하게 안알아두됨.
    if x.dim() == 1:
        x = x.view(int(x.size(0) ** .5), -1)
    
    plt.imshow(x, cmap='gray')
    plt.show()
```

```python
train_x, train_y = load_mnist(flatten=True) #train set 에 대해서 train_x(그림), train_y(label)를 가져오고 flatten(784차원으로) 펴라
test_x, test_y = load_mnist(is_train=False, flatten=True)

train_cnt = int(train_x.size(0) * config.train_ratio)
valid_cnt = train_x.size(0) - train_cnt

# Shuffle dataset to split into train/valid set.
indices = torch.randperm(train_x.size(0))
train_x, valid_x = torch.index_select(
    train_x,
    dim=0,
    index=indices
).split([train_cnt, valid_cnt], dim=0)
train_y, valid_y = torch.index_select(
    train_y,
    dim=0,
    index=indices
).split([train_cnt, valid_cnt], dim=0)

print("Train:", train_x.shape, train_y.shape)
print("Valid:", valid_x.shape, valid_y.shape)
print("Test:", test_x.shape, test_y.shape)
```

```python
from model import Autoencoder
```

```python
model = Autoencoder(btl_size=config.btl_size)
optimizer = optim.Adam(model.parameters())
crit = nn.MSELoss()

trainer = Trainer(model, optimizer, crit)
```

```python
trainer.train((train_x, train_x), (valid_x, valid_x), config) 
#y값인 train_x는 사실 필요없음. x값이 들어가면 x_hat을 뱉어낼꺼라서. y는 자기 자신이라서 그대로 train_x를 넣어준다.
#train_x을 넣은값과 train_x를 비교하도록 (train_x, train_x) 를 넣는거임
```

## Test

```python
with torch.no_grad():
    import random #random하게 골라달라고 가져옴

    index = int(random.random() * test_x.size(0)) #test set은 10000개. 10000개중에서 랜덤한 인덱스 뽑음

    recon = model(test_x[index].view(1, -1)).squeeze() #model를 거치면 (1,784)차원의 벡터가 나온다.

    #원래 test_x의 크기는 |test_x| = (10000, 784) 이고 |test_x[index]| 는 (784,)이므로
    #모델은 2차원 shape를 받을수있기에 애를 다시 (1,784)로 만들기위해 .view()를 쓴다
    #이걸 모델에 넣으면 모델은 똑같은 (1,784)를 뱉어내고 여기서 .squeeze하니깐 다시 (784,)가 된다
    
    
    show_image(test_x[index]) #test_x[index]는 원래 원본(784,)를 통과시키고
    show_image(recon) #비교하기위해 학습화된 recon도 통과시킨다
    
    #현재 bottleneck 사이즈가 2인데 굉장히 작아서 너무많이 압축을 해버려서 정보를 많이 버림
```

## Latent Space = Hidden layer의 결과물

각각의 hidden layer의 압축된 결과물이 다 latent space임

```python
Image("C:\\Users\\ryusu\\OneDrive\\바탕 화면\\5.png")
```

x_i번째 테스트셋(784차원)이 인코더를 통과하면 z_i(2차원)으로 나오게 된다

```python
if config.btl_size == 2:
    color_map = [
        'brown', 'red', 'orange', 'yellow', 'green',
        'blue', 'navy', 'purple', 'gray', 'black',
    ]
    #0번 인덱스는 0에 대한 색깔
    
    plt.figure(figsize=(20, 10))
    with torch.no_grad():
        latents = model.encoder(test_x[:1000]) #test set을 셔플링안하고 1000개만 통과 = test_x[:10000]. 그럼 z가 1000개나옴
        #그럼 latent의 크기는 |latent| = (1000,2) 의 텐서가 나옴
        
        
        for i in range(10): #우리는 for문을 돌면서 색깔별로 샘플을 찍어줄거임
            target_latents = latents[test_y[:1000] == i]
            target_y = test_y[:1000][test_y[:1000] == i]
            plt.scatter(target_latents[:, 0], #target_latent의 첫번째값을 x로 주고 
                        target_latents[:, 1], #target_latent의 두번째값을 y로 둬서 색깔을 주면서 scatter함
                        marker='o',
                        color=color_map[i],
                        label=i)
        
        plt.legend()
        plt.grid(axis='both')
        plt.show()
```

결과

숫자1은 빨간색이고
숫자4(초록색), 숫자9도 진한 녹색, 즉 뭉쳐있어서 아까 숫자4를 집어넣었는데 9가 뛰어나옴

```python
#위의 컬러점들이 많이 찍힌 좌표에서 범위를 정해서 디코더를 통과시켜 이미지 보기

if config.btl_size == 2: #bottleneck 사이즈가 2차원일때(원활하게 수행할수가 있어서)
    min_range, max_range = -2., 2. # -2부터 2까지 범위
    n = 20 #20번)개)을 샘플링
    step = (max_range - min_range) / float(n) # 좌표위 범위의 for문 정하기 :범위 -2 에서 2까지는 크기가 4인데 20번을 샘플링할꺼니깐 
                                                #4 나누기 20번하면 for 문의 step이 나옴
    
    with torch.no_grad(): #학습이 아니라 추론이라서 
        lines = []

        for v1 in np.arange(min_range, max_range, step): #min-range부터 max_range까지 step 띄는거,그게 z의 첫번째 좌표가 될거임
            #한줄씩 완성시킬거임. 첫윗줄은 -2 로 가득찬 floatTensor가 되는거임
            z = torch.stack([
                torch.FloatTensor([v1] * n),
                torch.FloatTensor([v2 for v2 in np.arange(min_range, #두번째줄은 4를 20으로 나눴으니깐 0.2이므로 [-2, -1.8... -1.8] 총 20개 floatTensor가 있을거임
                                                          max_range, step)]),
            ], dim=-1) #dim = -1은 [(20,a),(20,a)...]에서 a에 새로운 차원을 만들어 라는 뜻. 그럼 결과는 (20,2)가 나오게됨. 그럼 그게 z사이즈
            
            #v1의 결과물은 ([20,])이고 v2의 결과물은 [20,]이니깐 for문 한바퀴의 결과물은 ([20,][20,])이고 이 for문이
            #20번 돌면서 stack를 쌓을거임
            #dim = =1 은 빈칸에 새로운 차원을 만들어줘 라는 의미이므로 (20,2)라는 결과가 나오게됨. 그게 z의 사이즈가 된다
            # |z| = (20,2)
            
            #|z| = (20,2)
            
            line = torch.clamp(model.decoder(z).view(n, 28, 28), 0, 1) #a모델 디코더에 z를 통과시키면 (20, 784) 차원이 될거임. 
            #clamp는 어떤 범위로 다 쳐버러는거 찾아봐
            line = torch.cat([line[i] for i in range(n - 1, 0, -1)], dim=0)
            lines += [line]
            
        lines = torch.cat(lines, dim=-1)
        plt.figure(figsize=(20, 20))
        show_image(lines)
```
