# Binary Classification

## Load Dataset from sklearn

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['class'] = cancer.target

df.tail()
```

```python
df.columns
```

## Convert to PyTorch Tensor

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
data = torch.from_numpy(df.values).float()

data.shape
```

```python
x = data[:, :10]
y = data[:, -1:]

print(x.shape, y.shape)
```

```python
# Train / Valid / Test ratio
ratios = [.6, .2, .2]
```

```python
train_cnt = int(data.size(0) * ratios[0])
valid_cnt = int(data.size(0) * ratios[1])
test_cnt = data.size(0) - train_cnt - valid_cnt
cnts = [train_cnt, valid_cnt, test_cnt]

print("Train %d / Valid %d / Test %d samples." % (train_cnt, valid_cnt, test_cnt))
```

```python
indices = torch.randperm(data.size(0))

x = torch.index_select(x, dim=0, index=indices)
y = torch.index_select(y, dim=0, index=indices)

x = x.split(cnts, dim=0)
y = y.split(cnts, dim=0)

for x_i, y_i in zip(x, y):
    print(x_i.size(), y_i.size())
```

You don't need to shuffle them, if you have fixed train/valid/test set.

## Set Hyper Parameters

```python
n_epochs = 10000
batch_size = 128
print_interval = 500
early_stop = 100
```

## Get DataLoaders

```python
from torch.utils.data import Dataset, DataLoader
```

Dataset 은 샘플과 정답(label)을 저장하고, DataLoader 는 Dataset 을 샘플에 쉽게 접근할 수 있도록 순회 가능한 객체(iterable)로 감쌉니다.

```python
class CustomDataset(Dataset): #Dataset를 상속받는다
    
    def __init__(self, data, labels):
        self.data = data
        self.labels = labels
        
        super().__init__()
        
    def __len__(self):
        return len(self.data)
    
    #미니배치 만들기
    def __getitem__(self, idx): 
        return self.data[idx], self.labels[idx]
```

```python
train_loader = DataLoader( #x[0] = train, x[1] = valid, z[0] = test
    dataset=CustomDataset(x[0], y[0]), #CustomDataset(data, label)이 들어가니깐 x[0] = data, x[1] = labels
    batch_size=batch_size,
    shuffle=True, # train_loader는 꼭 셔플을 해줘야한다
)
valid_loader = DataLoader(
    dataset=CustomDataset(x[1], y[1]),
    batch_size=batch_size,
    shuffle=False, #나머지 set 에 대해서는 shuffle = False
)
test_loader = DataLoader(
    dataset=CustomDataset(x[2], y[2]),
    batch_size=batch_size,
    shuffle=False,
)

print("Train %d / Valid %d / Test %d samples." % (
    len(train_loader.dataset),
    len(valid_loader.dataset),
    len(test_loader.dataset),
))
```

## Build Model & Optimizer

```python
model = nn.Sequential( 
    nn.Linear(x[0].size(-1), 6), #x[0].size(-1)은 10차원이다
    nn.LeakyReLU(),
    nn.Linear(6, 5),
    nn.LeakyReLU(),
    nn.Linear(5, 4),
    nn.LeakyReLU(),
    nn.Linear(4, 3),
    nn.LeakyReLU(),
    nn.Linear(3, y[0].size(-1)),
    nn.Sigmoid(), #binary clasification이니깐 softmax대신 시그모이드 씀. 
)

model
```

```python
optimizer = optim.Adam(model.parameters())
```

## Train

```python
from copy import deepcopy

lowest_loss = np.inf
best_model = None

lowest_epoch = np.inf
```

```python
train_history, valid_history = [], []

for i in range(n_epochs):
    model.train()
    
    # We can remove following code block.
    # ------------------------------------------------------
    # indices = torch.randperm(x[0].size(0))
    # x_ = torch.index_select(x[0], dim=0, index=indices)
    # y_ = torch.index_select(y[0], dim=0, index=indices)
    #
    # x_ = x_.split(batch_size, dim=0)
    # y_ = y_.split(batch_size, dim=0)
    # ------------------------------------------------------
    # 이 과정은 dataloader 안에 있기때문에 안해도 된다.(train_loader)

    train_loss, valid_loss = 0, 0
    y_hat = []
    
    for x_i, y_i in train_loader: #한 epoch안 미니배치 내의 for 문
        y_hat_i = model(x_i)
        loss = F.binary_cross_entropy(y_hat_i, y_i)

        optimizer.zero_grad()
        loss.backward()

        optimizer.step()        
        train_loss += float(loss) # This is very important to prevent memory leak.
                                  # float 안하면 loss_i, loss_i1, loss_i2 ...를 따로따로 저장해서 메모리 낭비가 되기때문에 숫자로 더해준다

    train_loss = train_loss / len(train_loader)
        
    model.eval()
    with torch.no_grad():
        valid_loss = 0
        
        for x_i, y_i in valid_loader:
            y_hat_i = model(x_i)
            loss = F.binary_cross_entropy(y_hat_i, y_i)
            
            valid_loss += float(loss)
            
            y_hat += [y_hat_i]
            
    valid_loss = valid_loss / len(valid_loader)
    
    train_history += [train_loss]
    valid_history += [valid_loss]
        
    if (i + 1) % print_interval == 0:
        print('Epoch %d: train loss=%.4e  valid_loss=%.4e  lowest_loss=%.4e' % (
            i + 1,
            train_loss,
            valid_loss,
            lowest_loss,
        ))
        
    if valid_loss <= lowest_loss:
        lowest_loss = valid_loss
        lowest_epoch = i
        
        best_model = deepcopy(model.state_dict())
    else:
        if early_stop > 0 and lowest_epoch + early_stop < i + 1:
            print("There is no improvement during last %d epochs." % early_stop)
            break

print("The best validation loss from epoch %d: %.4e" % (lowest_epoch + 1, lowest_loss))
model.load_state_dict(best_model)
```

## Loss History

```python
plot_from = 2

plt.figure(figsize=(20, 10))
plt.grid(True)
plt.title("Train / Valid Loss History")
plt.plot(
    range(plot_from, len(train_history)), train_history[plot_from:],
    range(plot_from, len(valid_history)), valid_history[plot_from:],
)
plt.yscale('log')
plt.show()
```

## Let's see the result!

```python
test_loss = 0
y_hat = []

model.eval()
with torch.no_grad():
    for x_i, y_i in test_loader:
        y_hat_i = model(x_i)
        loss = F.binary_cross_entropy(y_hat_i, y_i)

        test_loss += loss # Gradient is already detached.

        y_hat += [y_hat_i]

test_loss = test_loss / len(test_loader)
y_hat = torch.cat(y_hat, dim=0) #y_hat이 리스트형태라서 쭉 붙여줌

print("Test loss: %.4e" % test_loss)
```

```python
correct_cnt = (y[2] == (y_hat > .5)).sum()  #y[2]에는 0또는 1이 들어가있음. y_hat > .5이면 1, 아니면 0 임. 그걸 sum해보면
                                            # 맞은 개수가 들어있음
total_cnt = float(y[2].size(0))#y[2].size(0)는 전체 test셋의 사이즈. 나눌거니깐 미리 float씌어줌

print('Test Accuracy: %.4f' % (correct_cnt / total_cnt))

#결과값의 의미는 이 테스트셋은 93% 이다.
```
