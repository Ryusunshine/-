## 실습

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
from utils import load_mnist
from trainer import Trainer
```

```python
def show_image(x): ##들어오는 tensor를 그림을 찍어주기(이미지 만들기) 위해 차원 변경
    if x.dim() == 1:
        x = x.view(int(x.size(0) ** .5), -1)
    
    plt.imshow(x, cmap='gray')
    plt.show()
```

```python
from argparse import Namespace

config = { #가독성있게 딕셔너리 형태로 만들어줌
    'train_ratio': .8,
    'batch_size': 256,
    'n_epochs': 50,
    'verbose': 1,
    'btl_size': 2
}

config = Namespace(**config)

print(config)
```

```python
# train, test 이미지를 불러오고 비율에 맞게 나눠주기(load_mnist, indices, split, cnt)

train_x, train_y = load_mnist(flatten=True) #train_x = 이미지, Train_y = 그 이미지의 label 불러오기
test_x, test_y = load_mnist(is_train=False, flatten=True) #test_x = 이미지, test_y = label 불러오기
#2차원 배열(28 x 28 픽셀)의 이미지를 28 * 28 = 784 픽셀의 1차원 배열로 변환합니다. 즉 이미지 픽셀의 행을 펼쳐서 일렬로 늘립니다

train_cnt = int(train_x.size(0) * config.train_ratio)
valid_cnt = train_x.size(0) - train_cnt

indices = torch.randperm(train_x.size(0))# train 데이터만 셔플링해주기. 그 차원에 맞게 train, 비교할 valid 샘플의 index select
train_x, valid_x = torch.index_select( #train 샘플과 valid 샘플을 각자 비율로 맞게 split하기
    train_x,
    dim=0,
    index=indices
).split([train_cnt, valid_cnt], dim=0)
train_y, valid_y = torch.index_select(
    train_y,
    dim=0,
    index=indices
).split([train_cnt, valid_cnt], dim=0)

print("Train:", train_x.shape, train_y.shape)
print("Valid:", valid_x.shape, valid_y.shape)
print("Test:", test_x.shape, test_y.shape)
```

```python
from model import Autoencoder
```

```python
#학습에 필요한 변수들 설정해주기

model = Autoencoder(btl_size=config.btl_size) #bottleneck 사이즈 2로 잡음
optimizer = optim.Adam(model.parameters())
crit = nn.MSELoss()

trainer = Trainer(model, optimizer, crit)
```

```python
trainer.train((train_x, train_x), (valid_x, valid_x), config)
```

## Mean value in each space

```python
with torch.no_grad():
    import random

    # 랜덤하게 인덱스 두개를 뽑는
    index1 = int(random.random() * test_x.size(0))
    index2 = int(random.random() * test_x.size(0))
    
    #인코더 통과
    z1 = model.encoder(test_x[index1].view(1, -1)) 
    z2 = model.encoder(test_x[index2].view(1, -1))
    # * test_x의 크기는 10000장의 784차원. 거기서 인덱스 하나를 고르면 앞에 1 날라가서 (784,)만 남는다. 
    #그 후, view를 했으니깐 결과적으로 (1, 784)차원이 된다. 그리고 인코더에 통과시키면 (1,784)에서 (1,2)가 된다

    #디코더 통과
    recon = model.decoder((z1 + z2) / 2).squeeze() #z1, z2는 encoder에 들어가서 나온 출력값이다.
    
    #* z1과 z2의 평균을 내고 디코더에 통과시키면 
    #(1,784)차원으로 다시 변환되고 squeeze를 했으니깐 1이 날라가고 (784,)만 남아 하나의 벡터가 된다. 이걸 recon에 담는다

    show_image(test_x[index1])
    show_image(test_x[index2])
    show_image((test_x[index1] + test_x[index2]) / 2) #그냥 input space(784차원)에서 평균때린거
    show_image(recon)
```
